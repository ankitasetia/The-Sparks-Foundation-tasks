**Task 1 : Prediction using Supervised Machine Learning**

Description: We have to predict the percentage of a student based on the no. of study hours.
Also we have to predict the score if a student studies for 9.25 hrs/ day.
Since it involves only two variables,this is a simple linear regreesion task.

**STEP-1 Import The Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

**STEP-2 Read the Data from Dataset**

```python
url="../input/student-data/student_scores%20-%20student_scores.csv"
Student_data = pd.read_csv(url)
```

```python
print(Student_data.shape)
Student_data.head()
```

```python
Student_data.describe()
```

**Check Missing Values**

```python

Student_data.info()
```

there is no missing value in Dataset

**STEP-3 Visualising Dataset**

```python
# Plotting the distribution of scores
Student_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage Score')  
plt.xlabel('No. of Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

From the graph we can predict a positive linear relation between the number of hours studied and percentage of score.

```python
# Check the correlation between No.of Hours and Percentage Scores
Student_data.corr()
```

This shows the positive correlation between No.of Hours and Percentage Scores


**STEP-4 Data Preparation**

```python
#divide the data by iloc function
X = Student_data.iloc[:,:-1].values
Y =  Student_data.iloc[:,1].values
```

**STEP-5 Model Training**

Split the data into training and testing sets and train the algorithm.

```python
#Train the model
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0) 
regressor = LinearRegression()  
regressor.fit(X_train.reshape(-1,1), Y_train) 
```

**STEP-6 Visualising the Model**

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line,color='red');
plt.xlabel('No. of Hours Studied',fontsize=15)
plt.ylabel('Scores in percentage',fontsize=15)
plt.show()
```

**STEP-7 Making Predictions**

```python
# Testing data
print(X_test)
# Model Prediction 
Y_pred = regressor.predict(X_test)
```

**STEP-8 Comparing Actual Result to Predicted Model Result**

```python
#  Actual vs Predicted
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
df
```

**STEP-9 Predict The Score for 9.25 Hours/Day**

```python
X = 9.25
predicted_score = regressor.predict([[X]])
```

```python
print("No of Hours = {}".format(X))
print("Predicted Score = {}".format(predicted_score[0]))
```

**PREDICTED SCORE FOR 9.25 HOURS/DAY IS 93.69173248737539%**

**STEP-10 Evaluating the Model**

```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(Y_test, Y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
```
